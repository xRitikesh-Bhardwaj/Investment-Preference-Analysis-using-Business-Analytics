**TASK 3 : DESCRIPTIVE STATISTICS & DATA ANALYSIS**

**Objective**

Perform statistical analysis on numerical variables and analyze investment preferences to understand customer behavior. This level provides valuable business insights through summary statistics and visualizations.

```python
# Identify Numerical Columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

print("Numerical Columns:\n")
print(numerical_columns)
```

```python
# Descriptive Statistics
df[numerical_columns].describe()
```

```python
# Mean of Numerical Columns
mean_values = df[numerical_columns].mean()

print("Mean Values\n")
print(mean_values)
```

```python
# Median of Numerical Columns
median_values = df[numerical_columns].median()

print("Median Values\n")
print(median_values)
```

```python
# Mode of Numerical Columns
mode_values = df[numerical_columns].mode().iloc[0]

print("Mode Values\n")
print(mode_values)
```

```python
# Standard Deviation
std_values = df[numerical_columns].std()

print("Standard Deviation\n")
print(std_values)
```

```python
# Variance
variance_values = df[numerical_columns].var()

print("Variance\n")
print(variance_values)
```

```python
# Minimum Values
minimum_values = df[numerical_columns].min()

print("Minimum Values\n")
print(minimum_values)
```

```python
# Maximum Values
maximum_values = df[numerical_columns].max()

print("Maximum Values\n")
print(maximum_values)
```

```python
# Range
range_values = maximum_values - minimum_values

print("Range\n")
print(range_values)
```

```python
# Statistical Summary Table
statistics = pd.DataFrame({

    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values,
    'Std Dev': std_values,
    'Variance': variance_values,
    'Minimum': minimum_values,
    'Maximum': maximum_values,
    'Range': range_values

})

statistics
```

```python
# Histogram of Numerical Columns
df[numerical_columns].hist(figsize=(12,8), bins=10)

plt.suptitle("Distribution of Numerical Variables", fontsize=16)

plt.tight_layout()

plt.show()
```

**Task 10: Correlation Analysis**

```python
# Select relevant columns
correlation_df = df[['age', 'Duration', 'Expect']]

correlation_df.head()
```

```python
# Convert Investment Duration into numerical values
duration_mapping = {
    'Less than 1 year': 1,
    '1-3 years': 2,
    '3-5 years': 3,
    'More than 5 years': 4
}

# Convert Expected Returns into numerical values
expect_mapping = {
    '20%-30%': 1,
    '30%-40%': 2,
    '40%-50%': 3
}

correlation_df['Duration'] = correlation_df['Duration'].map(duration_mapping)
correlation_df['Expect'] = correlation_df['Expect'].map(expect_mapping)

correlation_df.head()
```

```python
print(df['Duration'].unique())
print(df['Expect'].unique())
```

```python
correlation_matrix = correlation_df.corr()

print(correlation_matrix)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5,
    fmt='.2f'
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=correlation_df,
    x='age',
    y='Duration',
    s=80
)

plt.title("Age vs Investment Duration")

plt.xlabel("Age")

plt.ylabel("Investment Duration (Numeric)")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=correlation_df,
    x='age',
    y='Duration',
    s=80
)

plt.title("Age vs Investment Duration")

plt.xlabel("Age")

plt.ylabel("Investment Duration (Numeric)")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    data=correlation_df,
    x='Duration',
    y='Expect',
    s=80
)

plt.title("Investment Duration vs Expected Returns")

plt.xlabel("Investment Duration")

plt.ylabel("Expected Returns")

plt.show()
```

```python
print("""
BUSINESS INSIGHT

• The correlation matrix measures the strength and direction of relationships between age, investment duration, and expected returns.

• Positive correlation indicates that both variables increase together.

• Negative correlation indicates that one variable increases while the other decreases.

• Correlation values close to zero suggest little or no linear relationship.

• These findings help financial institutions understand investor behavior and design suitable investment products.
""")
```
